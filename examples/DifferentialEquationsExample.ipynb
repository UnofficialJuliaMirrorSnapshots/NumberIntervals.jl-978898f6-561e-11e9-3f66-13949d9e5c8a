```julia
using NumberIntervals, OrdinaryDiffEq
```

```julia
prob = ODEProblem((u, p, t) -> cos(t), zero(NumberInterval), (0., 10.))
```

```julia
sol = solve(prob, Tsit5(), internalnorm = (u, t) -> abs(u).hi)
```

```julia
using Plots
```

```julia
ts = 0:1e-2:10
plot(ts, mid.(sol.(ts)), label="numerical")
plot!(ts, sin.(ts), label="analytical")
```

```julia
plot(ts, radius.(sol.(ts)), label="numerical uncertainty")
```
